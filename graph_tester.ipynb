```python
import json
```

```python
graph = "stexpanded"
mappings_file = "./_input/mappings/" + graph + ".json"
triples_file = "./_input/triples/" + graph + ".triples"
```

```python
with open(mappings_file) as file:
    mappings = json.load(file)
    mappings_reversed = {str(v): k for k, v in mappings.items()}
```

```python
edge_dict = dict()
with open(triples_file, "r") as file:
    for line in file:
        edge = str(line.strip().split("###")[2])
        if edge not in edge_dict:
            edge_dict[str(edge)] = 1
        else:
            edge_dict[str(edge)] += 1

sorted_edge_dict = dict(sorted(edge_dict.items(), key=lambda item: item[1], reverse=True))
```

```python
for k, v in sorted_edge_dict.items():
    print(mappings_reversed[k], "-", v)
```

```python
duplicates_dict = dict()
with open(triples_file, "r") as file:
    for line in file:
        items = line.strip().split("###")
        if int(items[0]) < int(items[1]):
            edge = str(items[0]) + "###" + str(items[1])
        else:
            edge = str(items[1]) + "###" + str(items[0])
        if edge not in duplicates_dict:
            duplicates_dict[str(edge)] = 1
        else:
            duplicates_dict[str(edge)] += 1

sorted_duplicates_dict = dict(sorted(duplicates_dict.items(), key=lambda item: item[1], reverse=True))
```

```python
for k, v in sorted_duplicates_dict.items():
    print(k, "-", v)
    break
```

```python
node_dict = dict()
with open(triples_file, "r") as file:
    for line in file:
        items = line.strip().split("###")
        if items[0] in node_dict:
            node_dict[items[0]] += 1
        else:
            node_dict[items[0]] = 1
        if items[1] in node_dict:
            node_dict[items[1]] += 1
        else:
            node_dict[items[1]] = 1
print(len(node_dict))
```

```python
for k, v in node_dict.items():
    print(mappings_reversed[k], "-", v)
```
