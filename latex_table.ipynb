```python
import os
import json
```

```python
folders_path = "/Users/matevass/Documents/Projects/KGEntityMatching-Public/results_union"
folders = os.listdir(folders_path)
order = ["memoryalpha-memorybeta", "stexpanded-memoryalpha", "swg-starwars", "swtor-starwars"]
```

```python
table = ""
for folder in folders:
    if folder == ".DS_Store":
        continue
    line = folder
    with open(folders_path + "/" + folder + "/stats.json") as emf:
        scores = json.load(emf)
    # print(scores)
    for pair in order:
        # print(scores[pair])
        line += " & " + str(scores[pair][0]) + " & " + str(scores[pair][1]) + " & " + str(scores[pair][2])
    line += " \\\\"
    # print(line)
    table += line + "\n"
```
