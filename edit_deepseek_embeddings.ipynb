```python
import json
```

```python
graph = "stexpanded"
embeddings_file = "results/embeddings_deepseek_error/" + graph + "_summaries_v1_embeddings.json"
mappings_file = "./data/triples_v2/" + graph + "_mapping.json"
embeddings_output_file = "results/embeddings_deepseek/" + graph + "_summaries_v1_embeddings.json"
```

```python
with open(embeddings_file) as ef, open(mappings_file) as mf:
    embeddings = json.load(ef)
    mappings = json.load(mf)
```

```python
for k, v in embeddings.items():
    print(k)
    print(v)
    break
```

```python
for k, v in mappings.items():
    print(k)
    print(v)
    break
```

```python
embeddings = {mappings[k]: v for k, v in embeddings.items()}
```

```python
for k, v in embeddings.items():
    print(k)
    print(v)
    break
```

```python
with open(embeddings_output_file, "w") as eof:
    json.dump(embeddings, eof)
```
