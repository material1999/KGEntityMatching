```python
import json
```

```python
file = "/Users/matevass/Documents/Projects/KGEntityMatching-Public/results_union/dogtag_long_reranked10_deduplicated/memoryalpha-memorybeta.json"
```

```python
with open(file) as f:
    matches = json.load(f)
```

```python
from collections import Counter

left_urls = [pair[0] for pair in matches]
right_urls = [pair[1] for pair in matches]

left_freq = Counter(left_urls)
right_freq = Counter(right_urls)

left_gt1 = {url: count for url, count in left_freq.items() if count > 1}
right_gt1 = {url: count for url, count in right_freq.items() if count > 1}

left_sorted = dict(sorted(left_gt1.items(), key=lambda x: x[1], reverse=True))
right_sorted = dict(sorted(right_gt1.items(), key=lambda x: x[1], reverse=True))

print("Left frequencies > 1 (sorted):", left_sorted)
print("Right frequencies > 1 (sorted):", right_sorted)
```
