```python
import os
import re
import ast
```

```python
graph = "starwars"
communities_path = "./results/communities_leiden/" + graph + "/"
communities_final = communities_path + "final.txt"
hierarchy_file = communities_path + "hierarchy.txt"
hierarchy_file_edit = communities_path + "hierarchy_edit.txt"
```

```python
def get_highest_level(directory):
    max_level = -1
    pattern = re.compile(r'level_(\d+)\.txt')
    for filename in os.listdir(directory):
        match = pattern.match(filename)
        if match:
            level = int(match.group(1))
            max_level = max(max_level, level)
    return max_level if max_level != -1 else None

highest_level = get_highest_level(communities_path)
```

```python
with open(hierarchy_file_edit, "w") as hfe:
    line_list = []
    current_level = highest_level
    with open(hierarchy_file, "r") as hf:
        for line in hf:
            data = line.strip().replace("level ", "").split(" --- ")
            level_curr = int(data[0])
            comms_curr = ast.literal_eval(str(data[1]))
            # print(level_curr, comms_curr)
            line_list.append((level_curr, comms_curr))
            if len(comms_curr) == 1:
                hfe.write(line)
            else:
                print(line.strip())
                children = []
                for k, v in line_list:
                    if k == level_curr + 1:
                        if all(item in comms_curr for item in v):
                            children.append(v)
                print(str(children) + "\n")
                hfe.write("level " + str(level_curr) + " --- " + str(children) + "\n")
```
